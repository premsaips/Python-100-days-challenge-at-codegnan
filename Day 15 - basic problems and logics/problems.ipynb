```python
#find the len of the integer without len function

num = int(input())

def length(num):
    count = 0
    while num > 0:
        num //= 10
        count += 1
    return count
    
print(length(num))
    
```

```python
#find sum of digits

num = int(input())

def sum_num(num):
    sum = 0
    digit = 0
    while num > 0:
        digit = num % 10
        sum = sum + digit
        num //= 10
    return sum

print(sum_num(num))
```

```python
# Reverse a number or pallindrome of a number

num = int(input())

def rev(num):
    res = 0
    digit = 0
    while num > 0:
        digit = num % 10
        res = res * 10 + digit
        num //= 10
    return res

print(rev(num))
```
