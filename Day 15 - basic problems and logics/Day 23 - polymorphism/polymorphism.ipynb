# Polymorphism ---> poly means "many", morphism means "actions".

### ---> Here polymorphism means one method/function can perform different actions based on different inputs.


```python
# example

def add(x, y):
    return x + y

print(add(2, 3)) # ---> addition
print(add('code','gnan')) #---> concatenation
```

# Types of polymorphism
### 1. Runtime Polymorphism  ----> which occours while runtime {method overiding}
### 2. compile time Polymorphism  -----> which occurs while compiling {method overloading} [ex: syntax errors]

```python
#Method overriding
# Person
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        return f'I am {self.name} and i am {self.age}years old.'



# student
class student(person):
    def __init__(self, name, age, std):
        person.__init__(self, name, age)
        self.std = std

    def show(self):
        return f'I am {self.name} and i am studying {self.std} class. i am {self.age}years old'

p = person('prem',23)
s = student('sai', 15, 10)
print(p.show())
print(s.show())
```

```python
#Method overriding --> using super() function

# Person
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        return f'I am {self.name} and i am {self.age}years old.'



# student
class student(person):
    def __init__(self, name, age, std):
        person.__init__(self, name, age)
        self.std = std

    def show(self):
        print(super().show())
        return f'I am {self.name} and i am studying {self.std} class. i am {self.age}years old'

p = person('prem',23)
s = student('sai', 15, 10)
# print(p.show())
print(s.show())
```

# Method overloading
### ---> single class have multiple methods with same name with different parameter list.

```python
# method overloading

class math:
    def add(self,a,b):
        return a+b

    def add(self, a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d

m = math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
# method overloading
# by using default arguments

class math:

    def add(self,a,b,c = 0,d = 0):
        return a+b+c+d

m = math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
# method overloading
# by using variable len positional argument

class math:

    def add(self,a,b, *args):
        return a+b+ args

m = math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```
