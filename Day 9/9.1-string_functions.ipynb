# String Functions
## Case conversion functions
### 1. lower() --> changes all characters to small letter.
### 2. upper() --> changes all characters to captital letters.
### 3. title() --> changes start character to uppercase.
### 4. swapcase() --> convert upper to lower and lower to upper.
### 5. capitalize --> changes 1st character to uppercase.

```python
#lower
s = "I AM PREMSAI"
print(s.lower())
#upper()
s = "i am prem sai"
print(s.upper())
#title
s = "i am premsai"
print(s.title())
#swapcase()
s = "I am Premsai"
print(s.swapcase())
#capitalize()
s = "i am premsai"
print(s.capitalize())
```

## Sub string operations
### 1. find() --> Returns the index number of the characters 1st occurance
### 2. strip() --> Removes charecters / spaces in the string
### 3. split() --> Splits the string into parts

```python
#find()
s = "i am in codegnan"
print(s.find("t"))
print(s.find("c"))
print(s.find("n"))
```

```python
#strip()
s = "    codegnan   "
print(s.strip())
print(s.rstrip())
print(s.lstrip())
s = "@@@@@codegnan@@"
print(s.strip('@'))
```

```python
#split()
s = "@@ codegnan @"
print(s.split())
print(s.split('e'))
print(s.split('n'))
```

## case checking functions
### 1. isupper() --> Checks wheather the string is uppercase or not
### 2. islower() --> Checks wheather the string is lowercase or not
### 3. isdigit() --> Checks wheather the string is numbers or not
### 4. isalpha() --> Checks wheather the string is alphabets or not
### 5. isalnum() --> Checks wheather the string is character/numbers. return False for special characters

```python
#isupper()
s = "Code"
s1 = "CODE"
print(s.isupper())
print(s1.isupper())
```

```python
#islower()
s = "code"
s1 = "CODE"
print(s.islower())
print(s1.islower())
```

```python
#isdigit()
s = "C212"
s1 = "12345"
print(s.isdigit())
print(s1.isdigit())
```

```python
#isalpha()
s = "Code"
s1 = "COD1"
print(s.isalpha())
print(s1.isalpha())
```

```python
#isalnum()
s = "Co231"
s1 = "CODE@"
print(s.isalnum())
print(s1.isalnum())
```
