```python
#convert list of numbers into even list and odd list

lst1 = list(map(int, input().split()))
even = []
odd = []
for i in lst1:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
print("list:",lst1)
print("even",even)
print("odd",odd)
```

```python
#Marks analyze
#in listcontans list of students marks, in that find the pass percentage and fail percentage
#pass maeks is 35+
#fail is below 35

marks = list(map(int, input().split()))
pass_count = 0
fail_count = 0

for num in marks:
    if num >= 35:
        pass_count += 1
    else:
        fail_count += 1
pass_per = (pass_count/len(marks))*100
fail_per = (fail_count/len(marks))*100
print("pass % = ",pass_per)
print("fail % = ", fail_per)
```
