```python
#List functions
#1. --->append() = add single element to the end of the list
lst1 =[1]
lst2 = [2,3,4,6,5]
lst1.append(15)
lst1.append(lst2)
lst1.append(10)
print(lst1)
```

```python
#2. extend() = add multiple elements at the end

lst1 =[1]
lst2 = [2,3,4,6,5]
lst1.extend(lst2)
print(lst1)
```

```python
#3. insert() = insert element at position

lst1 =[1]
lst2 = [2,3,4,6,5]
lst1.insert(2, 100)
lst1.insert(2,200)
print(lst1)
lst1.insert(100,1000)
print(lst1)
```

```python
#4. pop() = dlete last element returnns the value
lst1 = [2,3,4,6,5]
lst1.pop()
lst1.pop(2)
print(lst1)
```

```python
#5. remove() = removes 1st occurance of the element
lst1 = [2,3,4,6,5]
lst1.remove(2)
lst1.remove(5)
print(lst1)
```

```python
#del = 
lst1 = [2,3,4,5,7,7,8]
del lst1[2]
print(lst1)
del lst1[2 : 4]
print(lst1)
```

```python
# count()
lst1 = [2,3,4,5,7,7,8]
print(lst1.count(5))
print(lst1.count(7))
print(lst1.count(10))
```

```python
# index()
lst1 = [2,3,4,5,7,7,8]
print(lst1.index(5))
print(lst1.index(7))
print(lst1.index(7,2))
#print(lst1.index(10))  -----Error
```

```python
#sort()
lst = [1,2,3,8,4,5,6,7]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```
