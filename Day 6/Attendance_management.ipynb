```python
n = int(input("Enter number of students: "))
present_count = 0
absent_count = 0
roll_no = 1
while roll_no <= n:
    print("enter roll no", roll_no, "present or absent. \n select 1. present  2. absent")
    status = input()
    if status == "1":
        present_count += 1
        roll_no += 1
    elif status == "2":
        absent_count += 1
        roll_no += 1
    else:
        print("select 1 or 2.")

print("Total noof students: ",n)
print("Noof students present: ", present_count)
print("Noof students absent: ", absent_count)
percentage = (present_count/n)*100
print("Attendance report: ",percentage,"%")
```
