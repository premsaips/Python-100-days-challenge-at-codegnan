```python
s = input()

val_str = "qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM1234567890@_"

#->
if len(s) >= 8:
    #Validate username
    #->First Character not a special character
    if s[0].isupper() or s[0].islower():
        for i in range(1, len(s)):
            #-> Username should cantains only alphanumeric and spl character like @_ 
            if s[i] not in val_str:
                print(f"Username is invalid bcz of {s[i]}")
                break
        else:
            print("Username is valid")
    else:
        print("Username invalid bcz first character should not be spl char")
        
else:
    print("User name is invalid!. It should contains minimum 8 characters")
```
